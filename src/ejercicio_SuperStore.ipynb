```python
region = df.groupby('Region')['Sales'].sum()
region
```

La región west es la que generó más **ventas** en este periodo

```python
region_profit = df.groupby('Region')['Profit'].sum()
region_profit
```

La región west es la que tuvo más **ganancia** en este periodo

```python
region = df.groupby('Region').agg({
    'Sales': 'sum',
    'Profit': 'sum',
    'Quantity': 'sum'
}).reset_index()

region
```

```python
# Calcular margen de ganancia  (ganancia/ventas) * 100
region['porcentaje_mg'] = (region['Profit'] / region['Sales']) * 100
region
```

```python
clientes_sales = df.groupby(['Region', 'Segment'])['Sales'].sum()
clientes_sales
```

```python
clientes_size = df.groupby(['Region', 'Segment']).size().reset_index(name='Count')
clientes_size
```

```python
total = df.groupby('Region').size().reset_index(name='Total')
total
```

```python
clientes = clientes_size.merge(total, on='Region')
clientes
```

```python
clientes['Porcentaje'] = ((clientes['Count'] / clientes['Total']) * 100).round(2)
clientes
```
